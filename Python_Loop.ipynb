### Introduction 
* In a loop, a part of a program is repeated over and over, until a specific goal is reached. 
* Loops are important for calculations that require repeated steps and for processing input consisting of many data items. In this chapter, you will learn about loop statements in Python, as well as techniques for writing programs that process input and simulate activities in the real world.

A loop can be used to tell a program to execute statements repeatedly. Suppose that you need to display a string (e.g., Programming is fun!) 100 times. It would be tedious to type the statement 100 times:

So, how do you solve this problem? Python provides a powerful construct called a loop, which controls how many times in succession an operation (or a sequence of operations) is performed. By using a loop statement, you don’t have to code the print statement a hundred times; you simply tell the computer to display a string that number of times. The loop statement can be written as follows:


```python
count = 0
while count<10:
    print( "I LOVE PYTHON")
    count=count+1
```

### 1. Python for Loop
In Python, the for loop is used to run a block of code for a certain number of times. It is used to iterate over any sequences such as list, tuple, string, etc.
<br>
A Python for loop iterates through each value in a sequence. Often you know exactly how many times the loop body needs to be executed, so a control variable can be used to count the executions. A loop of this type is called a counter-controlled loop.
* Syntax:<br>
`for iterating_var in sequence:
    statements(s)`

<br>
If a sequence contains an expression list, it is evaluated first. Then, the first item in the sequence is assigned to the iterating variable iterating_var. Next, the statements block is executed. Each item in the list is assigned to iterating_var, and the statement(s) block is executed until the entire sequence is exhausted.

### 2. Python while Loop

"while" loops<br>
While loops repeat as long as a certain boolean condition is met.<br>
A while loop executes statements repeatedly as long as a condition remains true. The syntax for the while loop is:<br>
`while loop-continuation-condition: <br>
#Loop body Statement(s)`
<br>

A single execution of a loop body is called an iteration (or repetition) of the loop. Each loop contains a loop-continuation-condition, a Boolean expression that controls the body’s execution. It is evaluated each time to determine if the loop body should be executed. If its evaluation is True, the loop body is executed; otherwise, the entire loop terminates and the program control turns to the statement that follows the while loop.

### 3. Loop Control Statements
The Loop control statements change the execution from its normal sequence. When the execution leaves a scope, all automatic objects that were created in that scope are destroyed. Python supports the following control statements.
* `Break statement:`
Terminates the loop statement and transfers execution to the statement immediately following the loop.
* `Continue statement:`
Causes the loop to skip the remainder of its body and immediately retest its condition prior to reiterating.
* `Pass statement:` 
The pass statement in Python is used when a statement is required syntactically but you do not want any command or code to execute.


#### 3.1. Break statement
The break statement is used for premature termination of the current loop. After abandoning the loop, execution at the next statement is resumed, just like the traditional break statement in C.
The most common use of break is when some external condition is triggered requiring a hasty exit from a loop. The break statement can be used in both while and for loops.
If you are using nested loops, the break statement stops the execution of the innermost loop and starts executing the next line of the code after the block.


```python
#Write a program to use break in for loop

for i in range (10):
    if i == 5 :
        break
    print(i)
    i=i+1
```

```python
#Write a program to use break in while loop

while i < 10:
    if i == 5 :
        break 
    print(i)
    i=i+1
```

#### 3.2. Continue Statement
The continue statement in Python returns the control to the beginning of the current loop. When encountered, the loop starts next iteration without executing the remaining statements in the current iteration.
The continue statement can be used in both while and for loops.


```python
#Write a program to use continue in for loop
for i in range (10):
    if i== 5:
        continue
    print(i)
    i=i+1
    
```

```python
#Write a program to use continue in for loop
while i < 10:
    if i== 5:
        continue
    print(i)
    i=i+1
```

### 4. Nested Loops
Python programming language allows to use one loop inside another loop.  A loop can be nested inside another loop.
Nested loops consist of an outer loop and one or more inner loops. Each time the outer loop is
repeated, the inner loops are reentered and started a new.

The syntax for a nested for loop statement in the Python programming language is as follows: <br>
`for iterator_var in sequence:
   for iterator_var in sequence:
       statements(s)
   statements(s)`

The syntax for a nested while loop statement in the Python programming language is as follows: <br>
`while expression:
   while expression: 
       statement(s)
   statement(s)`

`#outer for loop
for element in sequence 
   #inner for loop
   for element in sequence:
       body of inner for loop
   body of outer for loop`

```python
#Write a nested for loop program to print multiplication table in Python
for i in range (1,11):
    for j in range(1,11):
        print(i*j, end = " ")
    print()
```

```python
#Program to print half pyramid using *

for i in range (5):
    for j in range (5):
        print ("*",end= " ")
    print()
```

```python
for i in range (5):
    for j in range (i):
        print ("*",end= " ")
    print()
```

```python
#Program to print half pyramid a using numbers
rows = int (input (" ENTER THE NUMBER "))
for i in range (rows):
    for j in range(i):
        print (i, end=" ")
    print()
```

```python
#Inverted half pyramid using *
rows = int (input (" ENTER THE NUMBER "))
for i in range (rows,0,-1):
    for j in range(i):
        print (i, end=" ")
    print()
```

```python
#Inverted half pyramid using *
rows = int (input (" ENTER THE NUMBER "))
for i in range (rows,0,-1):
    for j in range(i):
        print ("*", end=" ")
    print()
```
